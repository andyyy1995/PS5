```python
import networkx as nx
import scipy as sp
from pylab import *
import numpy as np
from matplotlib import pyplot as plt
import networkx as nx
import seaborn as sns
import collections
from IPython.display import clear_output
import random
```

```python
graph = nx.readwrite.graphml.read_graphml('Wang2014.graphml')
```

```python
# Define drawing function (re-used many times later)
def circular_graph_drawing(G,figure_name):
    fig = figure(num=None, figsize=(10, 10), dpi=80)
    ax  = fig.add_subplot(111)
    nx.draw_networkx(G, pos=nx.drawing.layout.circular_layout(G), 
                     with_labels=True, font_weight='bold', ax=ax, node_size = 200)
    ax.set_axis_off()
    ax_lim = plt.axis('off') #by default, function operates on your "current" axis
    fig.savefig('./'+figure_name, bbox='tight')
```

```python
circular_graph_drawing(graph,'network_circular.pdf')
```
